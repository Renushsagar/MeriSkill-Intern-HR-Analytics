# Data Cleaning 

```python
#Deleting redundant columns.
columns_to_drop = ['DailyRate', 'EmployeeCount','JobInvolvement', 'JobLevel','MonthlyRate','Over18', 'PerformanceRating', 'RelationshipSatisfaction','StandardHours','StockOptionLevel','TrainingTimesLastYear','YearsWithCurrManager']
df = df.drop(columns=columns_to_drop)
df
```

```python
# Renaming the columns
df = df.rename(columns={'BusinessTravel': 'Business Travel', 'DistanceFromHome': 'Distance From Home', 'EducationField':'Education Field', 'EmployeeNumber':'Employee Number', 'EnvironmentSatisfaction':'Environment Satisfaction', 'HourlyRate':'Hourly Rate', 'JobRole':'Job Role', 'JobSatisfaction':'Job Satisfaction', 'MaritalStatus':'Marital Status','MonthlyIncome':'Monthly Income' , 'NumCompaniesWorked':'Num Companies Worked', 'PercentSalaryHike':'Percent Salary Hike',
       'TotalWorkingYears':'Total Working Years', 'WorkLifeBalance':'WorkLife Balance', 'YearsAtCompany':'Years At Company','YearsInCurrentRole':'Years In Current Role', 'YearsSinceLastPromotion':'Years Since Last Promotion'})
df
```

```python
#checking the duplicates
df.duplicated().any()
```

No Duplicates found.

```python
#Remove the NaN values from the dataset
df.dropna()
```

```python
# Checking for null values
df.isnull().any()
```

```python
# columns of data
df.columns
```

```python
print(df['Education'].unique())
```

```python
# Transforming the numerical values into Categorical values in Education column.
education_level = {
    1: 'Diploma',
    2: 'Graduate',
    3: 'Post Graduate',
    4: 'PhD',
    5: 'Others'
}

# Apply the mapping using replace
df['Education'] = df['Education'].replace(education_level)
df
```

# Data Visualization

```python
# Visualizing a correlation map for all numeric variables

corr = df.corr()
plt.figure(figsize=(30, 30))
sns.heatmap(corr, annot=True)
plt.show()
```

```python
# Visualizing OverTime 
ax = sns.countplot(x='OverTime', data=df)
# Add annotations
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2.0, p.get_height()), ha='center', va='bottom', fontsize=10)

plt.title('Overtime Distribution')
plt.show()
```

```python
# Visualizing Marital Status
ax = sns.countplot(x='Marital Status', data=df)
# Add annotations
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom', fontsize=10)

plt.title('Marital Status Distribution')
plt.show()
```

```python
# Visualizing Job Role
sns.countplot(y='Job Role', data=df)
plt.title('Job Role Distribution')
plt.show()
```

```python
# Visualizing Gender
ax = sns.countplot(x='Gender', data=df)
# Add annotations
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom', fontsize=10)

plt.title('Gender Distribution')
plt.show()
```

```python
# Visualizing Education Field 
sns.countplot(y='Education Field', data=df)
plt.title('Education Field Distribution')
plt.show()
```

```python
# Visualizing Department 
ax= sns.countplot(x='Department', data=df)
# Add annotations
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2.0, p.get_height()), ha='center', va='bottom', fontsize=10, color='black')

plt.title('Department Distribution')
plt.show()
```

```python
# Visualizing BusinessTravel 
ax = sns.countplot(x='Business Travel', data=df)

# Add annotations
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom', fontsize=10)

plt.title('BusinessTravel  Distribution')
plt.show()
```

```python
# Visualizing Relation between Overtime and Age
# Create the boxplot
sns.boxplot(x='OverTime', y='Age', data=df)
plt.title('Relation between Overtime and Age')
plt.show()
```

```python
# Visualizing Total Working Years
# Create the histogram
ax = sns.histplot(x='Total Working Years', bins=20, kde=True, data=df)

# Add labels to the bars
for p in ax.patches:
    x = p.get_x() + p.get_width()/2
    y = p.get_height()
    ax.annotate(f'{y:.0f}', (x, y), ha='center', va='bottom')

plt.title('Total Working Years')
plt.show()
```

```python
# Visualizing Education Level
ax=sns.countplot(x='Education', data=df)

# Add annotations
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2, p.get_height()), ha='center', va='bottom', fontsize=10)

plt.title('Education Level')
plt.show()
```
